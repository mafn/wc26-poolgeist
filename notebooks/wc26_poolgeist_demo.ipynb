# Poolgeist neutral demo

This notebook uses only neutral synthetic teams and fixtures. It does not contain private predictions or official tournament data. Poolgeist is for office-pool modelling and probabilistic experimentation, not betting advice.


```python
from poolgeist.models.ensemble import ModelCouncil
from poolgeist.reports.exports import export_prediction_bundle
from poolgeist.reports.summaries import summarize_prediction

council = ModelCouncil()
prediction = council.predict_match("Neutral A", "Neutral B")
summarize_prediction(prediction)
```

```python
prediction.score_ev_table.head(10)
```

```python
exported = export_prediction_bundle(prediction, "../outputs")
[str(path) for path in exported]
```
